# labelEncoder from scikit-learn (sklearn) is a useful tool for encoding target labels. It assigns a unique integer value to each label, making it easier to work with categorical data in machine learning models.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for col in sea_df.columns:
    if sea_df[col].dtype == 'object' or sea_df[col].dtype.name =='category':
        sea_df[col] = LabelEncoder().fit_transform(sea_df[col])
sea_df.head()
```

```python
# X and y column
X = sea_df.drop('survived', axis=1)
y = sea_df['survived']
```

```python
from sklearn.model_selection import train_test_split

```

```python
X_train,X_test ,y_train ,y_test = train_test_split(X,y ,test_size = 0.3,random_state = 0)
```

```python
X_train,X_test ,y_train ,y_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression().fit(X_train,y_train)
model
```

```python
y_pred_test = model.predict(X_test)
```

```python
y_pred_test
```

```python
y_pred_train = model.predict(X_train)
```

```python
y_pred_train
```

```python

plt.scatter(X_train,y_train,label = 'predict data')
plt.plot(X_train,y_train, color = 'brown',label = 'original data')

plt.show()
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix, classification_report
```

```python
accuracy = accuracy_score(y_test, y_pred_test)
recall = recall_score(y_test, y_pred_test)
precision = precision_score(y_test, y_pred_test)
f1 = f1_score(y_test, y_pred_test)
conf_matrix = confusion_matrix(y_test, y_pred_test)
class_report = classification_report(y_test, y_pred_test)
```
