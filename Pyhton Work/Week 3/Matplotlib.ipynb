#### sns.histplot(): Plots histogram.

# Pie Chart (Matplotlib):
#### plt.pie(): Plots pie charts.

# Scatter Plot (Matplotlib and Seaborn):
## Matplotlib:
#### plt.scatter(): Plots a scatter plot of x vs y.
## Seaborn:
#### sns.scatterplot(): Plots a scatter plot with additional options for styling and coloring.

# Histogram chart

```python
import matplotlib.pyplot as plt

population_ages=[22,33,44,33,22,55,66,77,88,87,98,33,22]
bins=[0,10,25,35,50,60,70,80,100,150]

plt.hist(population_ages,bins)
# HIST show the histogram 
```

```python
plt.hist(population_ages,bins, histtype='bar', rwidth=0.8, label='Ages')
# making histtype and wright 
```

```python
import matplotlib.pyplot as plt

population_ages=[22,33,44,33,22,55,66,77,88,87,98,33,22]
bins=[0,10,25,35,50,60,70,80,100,150]

plt.hist(population_ages,bins,histtype ='bar', rwidth=0.8, label='Ages')
# HIST show the histogram 

plt.xlabel('X axis')
plt.ylabel('y axis')
```

```python
import matplotlib.pyplot as plt

population_ages=[22,33,44,33,22,55,66,77,88,87,98,33,22]
bins=[0,10,25,35,50,60,70,80,100,150]

plt.hist(population_ages,bins,histtype ='bar', rwidth=0.8, label='Ages',color='green')
# HIST show the histogram 

plt.xlabel('X axis')
plt.ylabel('y axis')
plt.title('Data Visiulization')
plt.legend()
plt.show()
```

# Stackplot

```python
import matplotlib.pyplot as plt

days=[1,2,3,4,5]

reading=[3,6,8,9,9]
playing=[4,5,9,3,4]
watching=[13,7,6,8,12]
teaching=[2,15,20,4,5]

plt.plot([],[], color='c', label='days', linewidth=5)
plt.stackplot(days, reading,watching,playing,teaching, colors=['c'])

```
