```python

# Splitting variables
X = place_df.iloc[:, :1]  #as we know y axis is independent
y = place_df.iloc[:, 1:]  #as we know x axis is  dependent
```

```python
from sklearn.model_selection import train_test_split

# using the sklearn train and test funtion 

X_train , X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3 , random_state =  0)
X_train , X_test, y_train, y_test
```

```python
# Regressor model import linear regression liberry 
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

# or in other way we can use leaner regression model

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
reg
```

```python
# now we are going to Predict Results
y_pred_train = reg.predict(X_train)
y_pred_test  =reg.predict(X_test)
```

# now we are going the show or visuial the train model of linear regression

```python
plt.scatter(X_train,y_train,label = 'predict data')
plt.plot(X_train,y_train, color = 'brown',label = 'original data')
plt.title('Placement')
plt.xlabel('package')
plt.ylabel('cgpa')
plt.legend()

plt.show()
```

```python
plt.scatter(X_train,y_train,label = 'predict data')
plt.plot(X_train,y_train, color = 'brown',label = 'original data')
plt.title('Placement')
plt.xlabel('package')
plt.ylabel('cgpa')
plt.legend()

plt.box(False)
# Prediction on test set

plt.show()
```

```python
# Regressor coefficients and intercept
print(f'Coefficient: {reg.coef_}')
print(f'Intercept: {reg.intercept_}')
print('y = ', reg.intercept_, '+', reg.coef_, '* X')
```

```python
# evaluate the model
print('MSE = ', mean_squared_error(y_test, y_pred_test ))
print('R2 = ', r2_score(y_test, y_pred_test ))
print('RMSE = ', np.sqrt(mean_squared_error(y_test, y_pred_test )))
```
