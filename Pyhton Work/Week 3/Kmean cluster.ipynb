```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.cluster as KMeans
```

```python
sc = pd.read_csv('student_clustering.csv')
sc.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(sc.cgpa,sc.iq)
```

```python
wsc = [ ]
for i in range(1,21):
  Kmeans = KMeans.KMeans(n_clusters=i)
  Kmeans.fit(sc)
  wsc.append(Kmeans.inertia_)

```

```python
wsc
```

```python
plt.plot(range(1,21),wsc,marker='o',markeredgecolor = 'black')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
