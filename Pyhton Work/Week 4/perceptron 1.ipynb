```python
# Return weights and Intercept
def perceptron(x,y):

    X = np.insert(x,0,1,axis=1)
    weights = np.ones(x.shape[1])
    lr = 0.1

    for i in range(1000):
        j = np.random.randint(0,200)
        y_hat = step(np.dot(x[j],weights))# heree step is activtion FFuntion in this line we also use dot product
        weights = weights + lr*(y[j]-y_hat)*x[j]# difficient b/t actical value and predict value


    return weights[0],weights[1:]
```
